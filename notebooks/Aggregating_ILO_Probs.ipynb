```python
import numpy as np
import json

# 'P:\\FSE_UCM\\recommender systeem MSLAS\\course data\\augmented_course_data.json'
with open('augmented_course_data_v2.json', 'r') as file:
    data = json.load(file)
    blooms = dict()

    for (i, course) in enumerate(data):
        code = course['code']
        course_prob = dict()

        for (_, ilo_prob) in course['proba'].items():
            for (cls, p) in ilo_prob.items():
               course_prob[cls] = course_prob[cls] + [p] if cls in course_prob else [p]

        for (cls, prob) in course_prob.items():
            prob = np.array(prob)
            course_prob[cls] = 1 - np.prod((1 - prob))

        blooms[code] = course_prob

blooms
```

```python
# 'P:\\FSE_UCM\\recommender systeem MSLAS\\course data\\augmented_course_data_v2.json'
with open('precomputed_blooms.json', 'w') as file:
    json.dump(blooms, file)
```
