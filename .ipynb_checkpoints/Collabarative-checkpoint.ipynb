```python
from rec_sys_uni.datasets import datasets as ds
course_data = ds.get_course_data()
len(course_data)
```

```python
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
import numpy as np


# Convert 'appraisal_date' to datetime for comparison
student_data['appraisal_date'] = pd.to_datetime(student_data['appraisal_date'])

# Specify the target course
target_course = 'SCI2035'

# Get the appraisal dates for the target course
target_course_dates = student_data[student_data['module_abb'] == target_course].sort_values('appraisal_date', ascending=False).drop_duplicates('student_num')

# Merge this with the original data to compare dates
merged_data_with_dates = pd.merge(student_data, target_course_dates[['student_num', 'appraisal_date']], on='student_num', suffixes=('', '_target'))

# Filter out courses that were taken before the appraisal date of the target course
pre_target_courses = merged_data_with_dates[merged_data_with_dates['appraisal_date'] < merged_data_with_dates['appraisal_date_target']]

# Create a pivot table with grades for pre-target courses for each student
pivot_pre_target = pre_target_courses.pivot_table(index='student_num', columns='module_abb', values='grade', aggfunc='mean')

# Fill missing values with 0
pivot_pre_target.fillna(0, inplace=True)

# Adding missing courses to the pivot table
for course in course_data:
    if course not in pivot_pre_target.columns:
        pivot_pre_target[course] = 0

# Extract the target course grades
target_grades = student_data[student_data['module_abb'] == target_course].set_index('student_num')[['pass']]

# Join the pivot table with the target course grades
final_data_pre_target = pivot_pre_target.join(target_grades, how='inner')
final_data_pre_target = final_data_pre_target.drop(target_course, axis=1).reset_index()


# Separate features and target variable
X_pre_target = final_data_pre_target.drop('pass', axis=1)
y_pre_target = final_data_pre_target['pass']

# Initialize lists to store metrics
accuracies, precisions, recalls, f1s, roc_aucs = [], [], [], [], []

sss = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
for i, (train_index, test_index) in enumerate(sss.split(X_pre_target, y_pre_target)):
    X_train_pre, X_test_pre = X_pre_target.iloc[train_index], X_pre_target.iloc[test_index]
    y_train_pre, y_test_pre = y_pre_target.iloc[train_index], y_pre_target.iloc[test_index]
    # Train the Random Forest Regressor model with the pre-target course data
    model_pre = RandomForestClassifier(n_estimators=100, random_state=42)
    model_pre.fit(X_train_pre, y_train_pre)

    # Make predictions
    y_pred_pre = model_pre.predict(X_test_pre)

    # Calculate metrics
    accuracies.append(accuracy_score(y_test_pre, y_pred_pre))
    precisions.append(precision_score(y_test_pre, y_pred_pre))
    recalls.append(recall_score(y_test_pre, y_pred_pre))
    f1s.append(f1_score(y_test_pre, y_pred_pre))
    roc_aucs.append(roc_auc_score(y_test_pre, model_pre.predict_proba(X_test_pre)[:, 1]))

# Calculate averages and standard deviations
average_accuracy, std_accuracy = np.mean(accuracies), np.std(accuracies)
average_precision, std_precision = np.mean(precisions), np.std(precisions)
average_recall, std_recall = np.mean(recalls), np.std(recalls)
average_f1, std_f1 = np.mean(f1s), np.std(f1s)
average_roc_auc, std_roc_auc = np.mean(roc_aucs), np.std(roc_aucs)

# Print averages and standard deviations
print("Average Accuracy:", average_accuracy, "Std:", std_accuracy)
print("Average Precision:", average_precision, "Std:", std_precision)
print("Average Recall:", average_recall, "Std:", std_recall)
print("Average F1 Score:", average_f1, "Std:", std_f1)
print("Average ROC-AUC:", average_roc_auc, "Std:", std_roc_auc)
```

```python
# Extracting feature importances from the Random Forest model
feature_importances = pd.DataFrame(model_pre.feature_importances_, index=X_train_pre.columns, columns=['importance'])
# Sorting the features by importance
sorted_features = feature_importances.sort_values(by='importance', ascending=False)

# Calculating the correlation of these features with the target variable
correlation_with_target = X_train_pre.corrwith(y_train_pre)

# Combining the importance and correlation data
combined_feature_impact = pd.DataFrame({'Feature': sorted_features.index, 'Importance': sorted_features['importance']})
combined_feature_impact['Correlation with Grade'] = combined_feature_impact['Feature'].apply(lambda x: correlation_with_target[x])


combined_feature_impact
```
